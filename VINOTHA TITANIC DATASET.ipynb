Section 3: Data Cleaning

```python
#11.Replace the missing values in the Age column with the median age.
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age']
```

```python
#12.Replace missing values in the Embarked column with the most frequent value.
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked']
```

```python
#13.Drop the Cabin column from the dataset.
df=df.drop('Cabin',axis = 1)
df
```

```python
#14.Verify that the dataset no longer contains missing values in Age and Embarked.
df=df[['Age','Embarked']].isnull().sum()
df
```
